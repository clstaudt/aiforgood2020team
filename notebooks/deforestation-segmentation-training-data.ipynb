## Load Images

```python
region_coords = [5.727906, 51.993435, 5.588144, 51.944356]
```

```python
imgs = earthengine.earthengine2images(
    data_source="LANDSAT/LE07/C01/T1_SR",
    date_range=("2014-01-01", "2015-01-01"),
    cloud_cover=5,
    bands = {
        "R": "B3", 
        "G": "B2",
        "B": "B1"
    },
    region_coords=region_coords    
)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(imgs[0])
```
